```python
from sklearn.datasets import fetch_openml
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 #Fetching MNIST dataset in Scikit-Learn

mnist = fetch_openml('mnist_784', version=1,
cache=True)
X, y = mnist["data"], mnist["target"]
```

```python
X
```

```python
y
```

```python
#PCA
from sklearn.preprocessing import StandardScaler

X = StandardScaler().fit_transform(X)
pca = PCA(n_components=2)
pca_X = pca.fit_transform(X)
```

```python
pca_X 
```

```python
sns.scatterplot(x = pca_X[:,0], y = pca_X[:,1], hue = y, legend = 'full');
```

```python
#t_SNE
tsne = TSNE(n_components = 2, random_state = 42)
tsne_X = tsne.fit_transform(X)
```

```python
sns.scatterplot(x = tsne_X[:,0], y = tsne_X[:,1], hue = y, legend = 'full')
```

- TNSE chạy mất nhiều thời gian hơn PCA
- TSNE visuallize dữ liệu tốt hơn.
- TSNE cung cấp cái nhìn sâu sắc hơn về tập dữ liệu của bằng cách duy trì cấu trúc,sự tương quan của các điểm dữ liệu.
